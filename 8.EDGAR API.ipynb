```python
import requests
import json
import pandas as pd
```

```python
cik = ['0000803401','0001143311']
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrom/109.0.0.0 Safari/537.36'}
FactTable = []
Recent_Filings = []
Mailing_Address = []
Former_Names = []
Tickers = []
Exchanges = []

for x in cik:
    # Pull the Data
    url = f'https://data.sec.gov/submissions/CIK{x}.json'
    r = requests.get(url, headers = headers)
    data = r.json()

    # Structure Data
    df_1 = pd.json_normalize(data)
    df_1 = df_1[df_1.columns.drop(list(df_1.filter(regex='filings')))]
    df_1 = df_1[df_1.columns.drop(list(df_1.filter(regex='addresses')))]
    df_1 = df_1.drop(columns=['formerNames','tickers','exchanges'])
    df_2 = pd.DataFrame.from_dict(data['filings']['recent'])
    df_2.insert(0,'cik',df_1.iloc[0,0]) # Primary Key
    df_3 = pd.DataFrame.from_dict(data['addresses'], orient='index').reset_index().rename(columns={'index':'Type'})
    df_3.insert(0,'cik',df_1.iloc[0,0]) # Primary Key
    df_4 = pd.DataFrame.from_dict(data['formerNames'])
    df_4.insert(0,'cik',df_1.iloc[0,0]) # Primary Key
    df_5 = pd.DataFrame.from_dict(data['tickers'])
    df_5.insert(0,'cik',df_1.iloc[0,0]) # Primary Key
    df_6 = pd.DataFrame.from_dict(data['exchanges'])
    df_6.insert(0,'cik',df_1.iloc[0,0]) # Primary Key

    # Create Tables
    FactTable.append(df_1)
    Recent_Filings.append(df_2)
    Mailing_Address.append(df_3)
    Former_Names.append(df_4)
    Tickers.append(df_5)
    Exchanges.append(df_6)

FactTable= pd.concat(FactTable)
Recent_Filings= pd.concat(Recent_Filings)
Mailing_Address= pd.concat(Mailing_Address)
Former_Names= pd.concat(Former_Names)
Tickers= pd.concat(Tickers)
Exchanges= pd.concat(Exchanges)
```
